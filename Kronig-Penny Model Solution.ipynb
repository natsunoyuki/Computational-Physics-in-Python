```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

a = 1
h = 1
m = 1
K0 = 2 * np.pi / a
k = 0.2 * K0 * 0
N = 10000
factor = -np.linspace(0.1, 10, N)
K = np.arange(-100 * np.pi, 100 * np.pi + K0, K0)
y = np.zeros(N)
x = np.zeros(N)
Ek = 1 / (2 * m) * h**2 * (k-K)**2

def func(E, U0, Ek):
    V = U0 * np.sum(1 / (Ek - E)) + 1
    return V

for i in range(N):
    U0 = factor[i]

    result = fsolve(func, np.array([1]), args = (U0, Ek))
    y[i] = result
    x[i] = -1 / U0

plt.figure(figsize = (15, 5))
plt.plot(x, y)
plt.grid(True)
plt.show()
```
